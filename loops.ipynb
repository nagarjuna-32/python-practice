```python
# Outer loop for rows
for row in range(1, 4):
    # Inner loop for columns
    for col in range(1, 4):
        product = row * col
        print(f"{row} x {col} = {product}", end="\t")
    print()  # Move to a new line after each row
```

```python
items = ["apple", "banana", "cherry", "target_item", "date", "elderberry"]

for item in items:
    print(f"Checking: {item}")
    if item == "target_item":
        print("Found it! Stopping the search.")
        break  # Exits the loop entirely
```
