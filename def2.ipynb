```python
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    result = ""
    for char in text:
        if char not in vowels:
            result += char
    return result

print(remove_vowels("Python is Awesome"))
```

```python
def greet(name, message="Welcome back"):
    return f"Hello {name}, {message}!"

print(greet("Zane"))
print(greet("Zane", "Good Morning"))
```
