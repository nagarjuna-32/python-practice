```python
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Using it in a loop
temps_c = [0, 20, 30, 100]

for c in temps_c:
    f = celsius_to_fahrenheit(c)
    print(f"{c}°C is {f}°F")
```

```python
def get_evens(numbers):
    even_list = []
    for n in numbers:
        if n % 2 == 0:
            even_list.append(n)
    return even_list

# Testing
my_nums = [1, 2, 3, 4, 5, 6, 7, 8]
only_evens = get_evens(my_nums)
print(only_evens) # [2, 4, 6, 8]
```
