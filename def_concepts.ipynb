```python
def calculate_total(prices, tax_rate):
    total = 0
    for price in prices:
        total += price

    final_amount = total * (1 + tax_rate)
    return round(final_amount, 2)

# Using the function for different shopping carts
cart_1 = [19.99, 5.50, 10.00]
cart_2 = [100.00, 25.00]

print(f"Cart 1 Total: ${calculate_total(cart_1, 0.07)}") # 7% tax
print(f"Cart 2 Total: ${calculate_total(cart_2, 0.05)}") # 5% tax
```
