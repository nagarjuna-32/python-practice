```python
def generate_email(first, last):
    # .lower() ensures the email is professional
    email = f"{first.lower()}.{last.lower()}@company.com"
    return email

# Testing the function
user1 = generate_email("John", "Doe")
user2 = generate_email("Alice", "SMITH")

print(user1) # john.doe@company.com
print(user2) # alice.smith@company.com
```
