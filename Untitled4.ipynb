```python
items = ["apple", "banana", "cherry", "target_item", "date", "elderberry"]

for item in items:
    print(f"Checking: {item}")
    if item == "target_item":
        print("Found it! Stopping the search.")
        break  # Exits the loop entirely
```

```python
for num in range(1, 11):
    if num == 5:
        continue  # Skips the rest of the code in THIS iteration
    print(num)
```

```python
inventory = [10, 0, -2, 5, "GOLD", 3]

for item in inventory:
    if item == "GOLD":
        print("CRITICAL: Gold found. Locking system!")
        break

    # Check if it's a number before comparing
    if isinstance(item, int):
        if item < 0:
            print("Skipping error: Negative stock.")
            continue
        if item == 0:
            print("Alert: Item out of stock.")
        else:
            print(f"Item count: {item}")
```
